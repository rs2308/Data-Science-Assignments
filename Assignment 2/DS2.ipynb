```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('bitcoin prices.xlsx')
```

```python
sorted=data.sort_values(by=['Prices'],ascending=False)
```

```python
sorted
```

```python
prices=list()
# for i in range(10):#len(data['Prices'])
#     prices.append(sorted['Prices'][i])
prices = sorted['Prices'].tolist()
```

```python
prices
```

```python
Bitcoin=list()
# for i in range(10):#len(data['Bitcoin'])
#     Bitcoin.append(sorted['Bitcoin'][i])
Bitcoin = sorted['Bitcoin'].tolist()
```

```python
Bitcoin
```

```python
topPrices=list()
topBitcoin=list()
```

```python
for i in range(10):
    topPrices.append(prices[i])
    topBitcoin.append(Bitcoin[i])
```

```python
topPrices
```

```python
topBitcoin
```

```python
plt.xlabel('Price (in lakhs)')
plt.ylabel('Crypo Currency')
plt.barh(topBitcoin,topPrices)
plt.savefig('result.png')
```
