```python
file=open("data.txt")
original=file.read()
data=original.split(" ")
file.close()
```

```python
data
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
filtered_words = [word for word in data if word not in stopwords.words('english')]
```

```python
after_data=" ".join(filtered_words)
```

```python
frequent_data={}
for d in after_data.split():
    if d in frequent_data.keys():
        frequent_data[d]+=1
    else:
        frequent_data[d]=1
```

```python
frequent_data
```

```python
import operator

to_plot=sorted(frequent_data.items(),key=operator.itemgetter(1),reverse=True)[:20]
```

```python
to_plot
```

```python
import matplotlib.pyplot as plt
```

```python
x=[]
y=[]
for i in to_plot:
    x.append(i[0])
    y.append(i[1])
```

```python
x,y
```

```python
plt.figure(figsize=(20,20))
plt.barh(x,y)
plt.xlabel('Frequently used words')
plt.ylabel('No. of times used')
plt.savefig('output.png')
```
